# 1) O problema do XOR
- Mostre a equação de ajuste dos pesos durante o processo de treinamento

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import pickle # Serve para salvar o modelo treinado
```

```python
# Representação do problema XOR
# X: entradas (4 exemplos, 2 características)
# y: saídas (4 exemplos, 1 alvo)
x = np.array([
  [0, 0],
  [0, 1],
  [1, 0],
  [1, 1]
], dtype=np.float32)

y = np.array([
  [0],
  [1],
  [1],
  [0]
], dtype=np.float32)

# Visualização rápida
print("X:\n", x)
print("y:\n", y.ravel())
```

```python
#Importando o modelo da Rede Neural

from ModeloPy.Modelopy import RnModel
```

```python
hidden_neurons = 3 # 3 neurónios na camada intermediaria/oculta
output_neurons = 1 # 2 classes, ou seja 2 neurónios na camada de Saida
random_seed = 8 
learning_rate = 0.1 # taxa de aprendizado
epochs = 1000 # épocas de treinamento
```

```python
modelo = RnModel(
  

  
    x, y,
    hidden_neurons=hidden_neurons,
    output_neurons=output_neurons,
    random_seed=random_seed
)
result = modelo.fit(learning_rate=learning_rate, epochs=epochs)
```

```python
print("Predições finais:", result)
print("Saídas reais:", y.ravel())
```

```python
# Plotando histórico de perda
plt.plot(modelo.loss_history)
plt.title("Histórico de perda")
plt.xlabel("Épocas")
plt.ylabel("Loss")
plt.show()
```
